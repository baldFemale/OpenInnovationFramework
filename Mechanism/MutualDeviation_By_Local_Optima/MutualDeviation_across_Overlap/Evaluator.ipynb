```python
data_folder = r"F:\data\gst-0815-23\Mechanism_6_use_local_optima\MutualDeviation_across_Overlap_2"
overlap_list = [1, 2, 3]
gg_overlap_list = [3, 4, 5, 6]
gg_deviation_across_K_overlap, gs_deviation_across_K_overlap, ss_deviation_across_K_overlap, sg_deviation_across_K_overlap = [], [], [], []
for overlap in overlap_list:
    gs_deviation_file = data_folder + r"\gs_mutual_deviation_across_K_overlap_{0}".format(overlap)
    ss_deviation_file = data_folder + r"\ss_mutual_deviation_across_K_overlap_{0}".format(overlap)
    sg_deviation_file = data_folder + r"\sg_mutual_deviation_across_K_overlap_{0}".format(overlap)
    with open(gs_deviation_file, 'rb') as infile:
        gs_deviation = pickle.load(infile)
    with open(ss_deviation_file, 'rb') as infile:
        ss_deviation = pickle.load(infile)
    with open(sg_deviation_file, 'rb') as infile:
        sg_deviation = pickle.load(infile)
    gs_deviation_across_K_overlap.append(gs_deviation)
    ss_deviation_across_K_overlap.append(ss_deviation)
    sg_deviation_across_K_overlap.append(sg_deviation)

for gg_overlap in gg_overlap_list:
    gg_deviation_file = data_folder + r"\gg_mutual_deviation_across_K_overlap_{0}".format(gg_overlap)
    with open(gg_deviation_file, 'rb') as infile:
        gg_deviation = pickle.load(infile)
    gg_deviation_across_K_overlap.append(gg_deviation)

print(gg_deviation_across_K_overlap)
# print(np.array(ss_satisfaction_across_K_alpha).shape)
# print(gs_satisfaction_across_K_alpha)
# print(sg_satisfaction_across_K_alpha)
```

```python
# Integrate G and S
G_deviation_across_K_overlap = []
for gg_across_K, sg_across_K in zip(gg_deviation_across_K_overlap, sg_deviation_across_K_overlap):
    new_g_list = []
    for gg, sg in zip(gg_across_K, sg_across_K):
        new_g = 0.5 * (gg + sg)
        new_g_list.append(new_g)
    G_deviation_across_K_overlap.append(new_g_list)

S_deviation_across_K_overlap = []
for ss_across_K, gs_across_K in zip(ss_deviation_across_K_overlap, gs_deviation_across_K_overlap):
    new_s_list = []
    for ss, gs in zip(ss_across_K, gs_across_K):
        new_s = 0.5 * (ss + gs)
        new_s_list.append(new_s)
    S_deviation_across_K_overlap.append(new_s_list)
x = range(9)
fig, ax = plt.subplots()
ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)
# plt.plot(x, G_deviation_across_K_overlap[0], marker='o', linestyle='-', label="G Low Overlap", color=nature_blue)
# plt.plot(x, G_deviation_across_K_overlap[1], marker='s', linestyle='-', label="G Medium Overlap", color=nature_blue)
# plt.plot(x, G_deviation_across_K_overlap[2], marker='v', linestyle='-', label="G High Overlap", color=nature_blue)

plt.plot(x, G_deviation_across_K_overlap[0], linestyle='-', label="G Low Overlap", color=nature_blue)
plt.plot(x, G_deviation_across_K_overlap[1], linestyle='--', label="G Medium Overlap", color=nature_blue)
plt.plot(x, G_deviation_across_K_overlap[2], linestyle=':', label="G High Overlap", color=nature_blue)

plt.plot(x, S_deviation_across_K_overlap[0], marker='o', linestyle='-', label="S Low Overlap", color=nature_orange)
plt.plot(x, S_deviation_across_K_overlap[1], marker='s', linestyle='-', label="S Medium Overlap", color=nature_orange)
plt.plot(x, S_deviation_across_K_overlap[2], marker='v', linestyle='-', label="S High Overlap", color=nature_orange)

plt.xlabel('K', fontweight='bold', fontsize=12)
plt.ylabel('Mutual Deviation', fontweight='bold', fontsize=12)
plt.xticks(x)
plt.title("Knowledge Overlap")
plt.legend(frameon=False, fontsize=12)
plt.savefig(data_folder + r"\Mutual_Deviation_across_Overlap.png", transparent=True, dpi=300)
plt.show()
```
