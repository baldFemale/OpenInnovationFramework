```python
data_folder = r"E:\data\gst-0815-23\Mechanism_6_use_local_optima\MutualDeviation_across_Knowledge_2"
g_knowledge_list = [8, 12, 16]
s_knowledge_list = [8, 12, 16]
gg_deviation_across_K_knowledge, sg_deviation_across_K_knowledge = [], []
for knowledge in g_knowledge_list:
    gg_deviation_file = data_folder + r"\gg_mutual_deviation_across_K_knowledge_{0}".format(knowledge)
    sg_deviation_file = data_folder + r"\sg_mutual_deviation_across_K_knowledge_{0}".format(knowledge)
    with open(gg_deviation_file, 'rb') as infile:
        gg_deviation = pickle.load(infile)
    with open(sg_deviation_file, 'rb') as infile:
        sg_deviation = pickle.load(infile)
    gg_deviation_across_K_knowledge.append(gg_deviation)
    sg_deviation_across_K_knowledge.append(sg_deviation)

ss_deviation_across_K_knowledge, gs_deviation_across_K_knowledge = [], []
for knowledge in s_knowledge_list:
    ss_deviation_file = data_folder + r"\ss_mutual_deviation_across_K_knowledge_{0}".format(knowledge)
    gs_deviation_file = data_folder + r"\gs_mutual_deviation_across_K_knowledge_{0}".format(knowledge)
    with open(ss_deviation_file, 'rb') as infile:
        ss_deviation = pickle.load(infile)
    with open(gs_deviation_file, 'rb') as infile:
        gs_deviation = pickle.load(infile)
    ss_deviation_across_K_knowledge.append(ss_deviation)
    gs_deviation_across_K_knowledge.append(gs_deviation)

print(np.array(gg_deviation).shape)
print( gg_deviation_across_K_knowledge[0])
```

```python
# GST Fitness
x = range(9)
fig, ax = plt.subplots()
ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)
plt.plot(x, gg_deviation_across_K_knowledge[0], "-", label="GG Knowledge=12", color=nature_blue)
plt.plot(x, gg_deviation_across_K_knowledge[1], "-", label="GG Knowledge=14", color=nature_green)
plt.plot(x, gg_deviation_across_K_knowledge[2], "-", label="GG Knowledge=16", color=nature_orange)

# The solid line is for homogeneous components, so basically it is consistent with the independent perspective (without fitness and only by chance)
# However, when they are exposed to difference resource. They behave in totally different directions
# For S: S gains more when it searches upon the established solution from G
# For G: G gains less when it searches upon the established solutions from S. In other words, G also cannot get rid of the partial attractiveness established by S

plt.xlabel('K', fontweight='bold', fontsize=12)
plt.ylabel('Mutual Deviation', fontweight='bold', fontsize=12)
plt.xticks(x)
plt.legend(frameon=False, ncol=1, fontsize=12)
plt.savefig(data_folder + r"\Mutual_Deviation_across_Knowledge_GG.png", transparent=True, dpi=300)
plt.show()
```

```python
# Integrate G and S
x = range(9)
G_deviation_across_knowledge = []
for gg_across_K, sg_across_K in zip(gg_deviation_across_K_knowledge, sg_deviation_across_K_knowledge):
    new_g_list = []
    for gg, sg in zip(gg_across_K, sg_across_K):
        new_g = 0.5 * (gg + sg)
        new_g_list.append(new_g)
    G_deviation_across_knowledge.append(new_g_list)
S_deviation_across_knowledge = []
for ss_across_K, gs_across_K in zip(ss_deviation_across_K_knowledge, gs_deviation_across_K_knowledge):
    new_s_list = []
    for ss, gs in zip(ss_across_K, gs_across_K):
        new_s = 0.5 * (ss + gs)
        new_s_list.append(new_s)
    S_deviation_across_knowledge.append(new_s_list)

fig, ax = plt.subplots()
ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)
plt.plot(x, G_deviation_across_knowledge[0], marker='o', linestyle='-', label="G12", color=nature_blue)
plt.plot(x, G_deviation_across_knowledge[1], marker='s', linestyle='-', label="G14", color=nature_blue)
plt.plot(x, G_deviation_across_knowledge[2], marker='v', linestyle='-', label="G16", color=nature_blue)

plt.plot(x, S_deviation_across_knowledge[0], marker='o', linestyle='-', label="S12", color=nature_orange)
plt.plot(x, S_deviation_across_knowledge[1], marker='s', linestyle='-', label="S16", color=nature_orange)
plt.plot(x, S_deviation_across_knowledge[2], marker='v', linestyle='-', label="S20", color=nature_orange)
plt.xlabel('K', fontweight='bold', fontsize=12)
plt.ylabel('Mutual Deviation', fontweight='bold', fontsize=12)
plt.xticks(x)
plt.title("(b) Knowledge Amount")
plt.legend(frameon=False, ncol=2, fontsize=12)
plt.savefig(data_folder + r"\Mutual_Deviation_across_Knowledge.png", transparent=True, dpi=300)
plt.show()
```
