```python
data_folder = r"E:\data\gst-0815-23\Mechanism_5_use_satisfying\JointSatisfaction_across_Overlap_1"
overlap_list = [0, 1, 2, 3]
gg_satisfaction_across_K_overlap, gs_satisfaction_across_K_overlap, ss_satisfaction_across_K_overlap, sg_satisfaction_across_K_overlap = [], [], [], []
for overlap in overlap_list:
    gg_satisfaction_file = data_folder + r"\gg_joint_satisfaction_across_K_overlap_{0}".format(overlap)
    gs_satisfaction_file = data_folder + r"\gs_joint_satisfaction_across_K_overlap_{0}".format(overlap)
    ss_satisfaction_file = data_folder + r"\ss_joint_satisfaction_across_K_overlap_{0}".format(overlap)
    sg_satisfaction_file = data_folder + r"\sg_joint_satisfaction_across_K_overlap_{0}".format(overlap)
    
    with open(gg_satisfaction_file, 'rb') as infile:
        gg_satisfaction = pickle.load(infile)
    with open(gs_satisfaction_file, 'rb') as infile:
        gs_satisfaction = pickle.load(infile)
    with open(ss_satisfaction_file, 'rb') as infile:
        ss_satisfaction = pickle.load(infile)
    with open(sg_satisfaction_file, 'rb') as infile:
        sg_satisfaction = pickle.load(infile)
        
    gg_satisfaction_across_K_overlap.append(gg_satisfaction)
    gs_satisfaction_across_K_overlap.append(gs_satisfaction)
    ss_satisfaction_across_K_overlap.append(ss_satisfaction)
    sg_satisfaction_across_K_overlap.append(sg_satisfaction)
    print(sg_satisfaction)
# print(np.array(ss_satisfaction_across_K_alpha).shape)
# print(gs_satisfaction_across_K_alpha)
# print(sg_satisfaction_across_K_alpha)
```
