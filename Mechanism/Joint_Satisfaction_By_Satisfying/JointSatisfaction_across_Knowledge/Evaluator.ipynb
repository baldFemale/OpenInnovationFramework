```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
legend_properties = {'weight':'bold'}
nus_blue = "#003D7C"
nus_orange = "#EF7C00"
# Nature three colors
nature_orange = "#F16C23"
nature_blue = "#2B6A99"
nature_green = "#1B7C3D"
# Morandi six colors
morandi_blue = "#046586"
morandi_green =  "#28A9A1"
morandi_yellow = "#C9A77C"
morandi_orange = "#F4A016"
morandi_pink = "#F6BBC6"
morandi_red = "#E71F19"
morandi_purple = "#B08BEB"
# Others
shallow_grey = "#D3D4D3"
deep_grey = "#A6ABB6"
# Red and Blue
strong_red = "#A61C5D"
strong_blue = "#1B64A4"

# Shallow-deep pair
shallow_purple = "#EAD7EA"
deep_purple = "#BA9DB9"
shallow_cyan = "#A9D5E0"
deep_cyan = "#48C0BF"
shallow_blue = "#B6DAEC"
deep_blue = "#98CFE4"
shallow_pink = "#F5E0E5"
deep_pink = "#E5A7B6"
shallow_green = "#C2DED0"
deep_green = "#A5C6B1"

color_pairs = [[shallow_purple, deep_purple],  [shallow_green, deep_green],
               [shallow_blue, deep_blue], [shallow_pink, deep_pink], [shallow_cyan, deep_cyan],
               [shallow_grey, deep_grey]]
morandi_colors = [morandi_blue, morandi_green, morandi_yellow, morandi_orange, morandi_pink, morandi_red, morandi_purple]
```

```python
data_folder = r"F:\data\gst-0815-23\Mechanism_7_use_satisfying\Joint_Satisfaction_By_Satisfying\JointSatisfaction_across_Knowledge"
# GG
generalist_expertise_list =  [12, 14, 16, 18]
gg_satisfaction_across_knowledge, sg_satisfaction_across_knowledge = [], []
for g_expertise in generalist_expertise_list:
    gg_confusion_file = data_folder + r"\gg_joint_satisfaction_across_K_G_{0}".format(g_expertise)
    sg_confusion_file = data_folder + r"\sg_joint_satisfaction_across_K_G_{0}".format(g_expertise)
    with open(gg_confusion_file, 'rb') as infile:
        gg_confusion = pickle.load(infile)
    with open(sg_confusion_file, 'rb') as infile:
        sg_confusion = pickle.load(infile)
    # print(g_expertise, gg_confusion)
    gg_satisfaction_across_knowledge.append(gg_confusion)
    sg_satisfaction_across_knowledge.append(gg_confusion)

x = range(1, 9)
fig, ax = plt.subplots()
ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)
plt.plot(x, gg_satisfaction_across_knowledge[0][1:], "-", label="G10", color=nature_blue)
plt.plot(x, gg_satisfaction_across_knowledge[1][1:], "-", label="G12", color=nature_green)
plt.plot(x, gg_satisfaction_across_knowledge[2][1:], "-", label="G14", color=nature_orange)
plt.plot(x, gg_satisfaction_across_knowledge[3][1:], "-", label="G16", color=morandi_red)
plt.xlabel('K', fontweight='bold', fontsize=12)
plt.ylabel('Joint Satisfaction', fontweight='bold', fontsize=12)
plt.title("GG")
plt.xticks(x)
plt.legend(frameon=False, ncol=1, fontsize=12)
plt.savefig(data_folder + r"\Joint_Satisfaction_across_Knowledge_GG.png", transparent=True, dpi=300)
plt.show()
```

```python
# Integrate G and S
x = range(9)
G_satisfaction_across_knowledge = []
for gg_across_K, sg_across_K in zip(gg_satisfaction_across_knowledge, sg_satisfaction_across_knowledge):
    new_g_list = []
    for gg, sg in zip(gg_across_K, sg_across_K):
        new_g = 0.5 * (gg + sg)
        new_g_list.append(new_g)
    G_satisfaction_across_knowledge.append(new_g_list)
S_satisfaction_across_knowledge = []
for ss_across_K, gs_across_K in zip(ss_satisfaction_across_knowledge, gs_satisfaction_across_knowledge):
    new_s_list = []
    for ss, gs in zip(ss_across_K, gs_across_K):
        new_s = 0.5 * (ss + gs)
        new_s_list.append(new_s)
    S_satisfaction_across_knowledge.append(new_s_list)
      
fig, ax = plt.subplots()
ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)
plt.plot(x, G_satisfaction_across_knowledge[1], marker='o', linestyle='-', label="G12", color=nature_blue)
plt.plot(x, G_satisfaction_across_knowledge[2], marker='s', linestyle='-', label="G14", color=nature_blue)
plt.plot(x, G_satisfaction_across_knowledge[3], marker='v', linestyle='-', label="G16", color=nature_blue)

plt.plot(x, S_satisfaction_across_knowledge[0], marker='o', linestyle='-', label="S12", color=nature_orange)
plt.plot(x, S_satisfaction_across_knowledge[1], marker='s', linestyle='-', label="S16", color=nature_orange)
plt.plot(x, S_satisfaction_across_knowledge[2], marker='v', linestyle='-', label="S20", color=nature_orange)
plt.xlabel('K', fontweight='bold', fontsize=12)
plt.ylabel('Joint Confirmation', fontweight='bold', fontsize=12)
plt.xticks(x)
plt.title("Knowledge Amount")
plt.legend(frameon=False, ncol=2, fontsize=12)
plt.savefig(data_folder + r"\Joint_Satisfaction_across_Knowledge.png", transparent=True, dpi=300)
plt.show()
```
