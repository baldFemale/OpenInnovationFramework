```python
data_folder = r"E:\data\gst-0815-23\N9E12\EvaluatingKnowledge"
# GG
G_crowd_knowledge_list = [12, 14, 16, 18]
S_crowd_knowledge_list = [12, 16, 20, 24, 32, 36]

gg_average_across_E, gg_best_across_E, gg_variance_across_E, gg_diversity_across_E = [], [], [], []
for G_crowd_knowledge in G_crowd_knowledge_list:
    gg_performance_file = data_folder + r"\gg_performance_across_K_{0}".format(G_crowd_knowledge)
    gg_best_performance_file = data_folder + r"\gg_best_performance_across_K_{0}".format(G_crowd_knowledge)
    gg_variance_file = data_folder + r"\gg_variance_across_K_{0}".format(G_crowd_knowledge)
    gg_diversity_file = data_folder + r"\gg_diversity_across_K_{0}".format(G_crowd_knowledge)
    with open(gg_performance_file, 'rb') as infile:
        gg_average = pickle.load(infile)
    with open(gg_best_performance_file, 'rb') as infile:
        gg_best = pickle.load(infile)
    with open(gg_variance_file, 'rb') as infile:
        gg_variance = pickle.load(infile)
    with open(gg_diversity_file, 'rb') as infile:
        gg_diversity = pickle.load(infile)
    gg_average_across_E.append(gg_average)
    gg_best_across_E.append(gg_best)
    gg_variance_across_E.append(gg_variance)
    gg_diversity_across_E.append(gg_diversity)
    
# GS
gs_average_across_E, gs_best_across_E, gs_variance_across_E, gs_diversity_across_E = [], [], [], []
for G_crowd_knowledge in G_crowd_knowledge_list:
    gs_performance_file = data_folder + r"\gs_performance_across_K_{0}".format(G_crowd_knowledge)
    gs_best_performance_file = data_folder + r"\gs_best_performance_across_K_{0}".format(G_crowd_knowledge)
    gs_variance_file = data_folder + r"\gs_variance_across_K_{0}".format(G_crowd_knowledge)
    gs_diversity_file = data_folder + r"\gs_diversity_across_K_{0}".format(G_crowd_knowledge)
    with open(gs_performance_file, 'rb') as infile:
        gs_average = pickle.load(infile)
    with open(gs_best_performance_file, 'rb') as infile:
        gs_best = pickle.load(infile)
    with open(gs_variance_file, 'rb') as infile:
        gs_variance = pickle.load(infile)
    with open(gs_diversity_file, 'rb') as infile:
        gs_diversity = pickle.load(infile)
    gs_average_across_E.append(gs_average)
    gs_best_across_E.append(gs_best)
    gs_variance_across_E.append(gs_variance)
    gs_diversity_across_E.append(gs_diversity)

# SS
ss_average_across_E, ss_best_across_E, ss_variance_across_E, ss_diversity_across_E = [], [], [], []
for S_crowd_knowledge in S_crowd_knowledge_list:
    ss_performance_file = data_folder + r"\ss_performance_across_K_{0}".format(S_crowd_knowledge)
    ss_best_performance_file = data_folder + r"\ss_best_performance_across_K_{0}".format(S_crowd_knowledge)
    ss_variance_file = data_folder + r"\ss_variance_across_K_{0}".format(S_crowd_knowledge)
    ss_diversity_file = data_folder + r"\ss_diversity_across_K_{0}".format(S_crowd_knowledge)
    with open(ss_performance_file, 'rb') as infile:
        ss_average = pickle.load(infile)
    with open(ss_best_performance_file, 'rb') as infile:
        ss_best = pickle.load(infile)
    with open(ss_variance_file, 'rb') as infile:
        ss_variance = pickle.load(infile)
    with open(ss_diversity_file, 'rb') as infile:
        ss_diversity = pickle.load(infile)
    ss_average_across_E.append(ss_average)
    ss_best_across_E.append(ss_best)
    ss_variance_across_E.append(ss_variance)
    ss_diversity_across_E.append(ss_diversity)
# SG
sg_average_across_E, sg_best_across_E, sg_variance_across_E, sg_diversity_across_E = [], [], [], []
for S_crowd_knowledge in S_crowd_knowledge_list:
    sg_performance_file = data_folder + r"\sg_performance_across_K_{0}".format(S_crowd_knowledge)
    sg_best_performance_file = data_folder + r"\sg_best_performance_across_K_{0}".format(S_crowd_knowledge)
    sg_variance_file = data_folder + r"\sg_variance_across_K_{0}".format(S_crowd_knowledge)
    sg_diversity_file = data_folder + r"\sg_diversity_across_K_{0}".format(S_crowd_knowledge)
    with open(sg_performance_file, 'rb') as infile:
        sg_average = pickle.load(infile)
    with open(sg_best_performance_file, 'rb') as infile:
        sg_best = pickle.load(infile)
    with open(sg_variance_file, 'rb') as infile:
        sg_variance = pickle.load(infile)
    with open(sg_diversity_file, 'rb') as infile:
        sg_diversity = pickle.load(infile)
    sg_average_across_E.append(sg_average)
    sg_best_across_E.append(sg_best)
    sg_variance_across_E.append(sg_variance)
    sg_diversity_across_E.append(sg_diversity)
```
