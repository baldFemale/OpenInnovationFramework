```python
import matplotlib.pyplot as plt
import pickle
import numpy as np
```

```python
data_folder = r"E:\data\gst-1112\Composition_2\GST_mid"
g_performance_file = data_folder + r"\g_performance_across_K"
s_performance_file = data_folder + r"\s_performance_across_K"
# Potential Performamce
g_potential_file = data_folder + r"\g_potential_across_K"
with open(g_potential_file, 'rb') as infile:
    g_potential = pickle.load(infile)
with open(g_performance_file, 'rb') as infile:
    g_performance = pickle.load(infile)
with open(s_performance_file, 'rb') as infile:
    s_performance = pickle.load(infile)
```

```python
x = [0, 1, 2, 3, 4, 5, 6, 7, 8]
plt.plot(x, g_performance, "k--", label="Generalist")
plt.plot(x, g_potential, "k:", label="G Potential")
plt.plot(x, s_performance, "k-", label="Specialist")
# plt.plot(x, t_performance, "g-", label="T")
# plt.title('Diversity Decrease')
plt.xlabel('K', fontweight='bold', fontsize=10)
plt.ylabel('Performance', fontweight='bold', fontsize=10)
plt.xticks(x)
plt.title("(a)")
plt.yticks(np.arange(0.55, 1.05, step=0.05))
plt.legend(frameon=False)
plt.savefig(data_folder + r"\GS_performance_mid.png", transparent=True, dpi=200)
plt.show()
plt.clf()
```

```python
g_original_performance_file = data_folder + r"\g_original_performance_across_K"
s_original_performance_file = data_folder + r"\s_original_performance_across_K"
# t_original_performance_file = data_folder + r"\t_original_performance_across_K"
with open(g_original_performance_file, 'rb') as infile:
    g_original_performance = pickle.load(infile)
with open(s_original_performance_file, 'rb') as infile:
    s_original_performance = pickle.load(infile)
# with open(t_original_performance_file, 'rb') as infile:
#     t_original_performance = pickle.load(infile)
g_deviation, s_deviation = [], []
for g_K, s_K in zip(g_original_performance, s_original_performance):
    g_deviation.append(np.std(g_K))
    s_deviation.append(np.std(s_K))
print(np.array(g_original_performance).shape)
print(np.array(g_original_performance[0]).shape)
```
