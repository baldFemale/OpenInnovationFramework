```python
data_folder = r"/Volumes/T7/data/gst-0112-26/Mechanism_2/Tension_across_knowledge"
# G
gg_confirmation_file = data_folder + r"/gg_joint_confirmation_across_knowledge_K"
gg_deviation_file = data_folder + r"/gg_mutual_deviation_across_knowledge_K"
sg_confirmation_file = data_folder + r"/sg_joint_confirmation_across_knowledge_K"
sg_deviation_file = data_folder + r"/sg_mutual_deviation_across_knowledge_K"
with open(gg_confirmation_file, 'rb') as infile:
    gg_confirmation = pickle.load(infile)
with open(sg_confirmation_file, 'rb') as infile:
    sg_confirmation = pickle.load(infile)
with open(gg_deviation_file, 'rb') as infile:
    gg_deviation = pickle.load(infile)
with open(sg_deviation_file, 'rb') as infile:
    sg_deviation = pickle.load(infile)

# S
ss_confirmation_file = data_folder + r"/ss_joint_confirmation_across_knowledge_K"
ss_deviation_file = data_folder + r"/ss_mutual_deviation_across_knowledge_K"
gs_confirmation_file = data_folder + r"/gs_joint_confirmation_across_knowledge_K"
gs_deviation_file = data_folder + r"/gs_mutual_deviation_across_knowledge_K"
with open(ss_confirmation_file, 'rb') as infile:
    ss_confirmation = pickle.load(infile)
with open(ss_deviation_file, 'rb') as infile:
    ss_deviation = pickle.load(infile)
with open(gs_confirmation_file, 'rb') as infile:
    gs_confirmation = pickle.load(infile)
with open(gs_deviation_file, 'rb') as infile:
    gs_deviation = pickle.load(infile)
print(np.shape(gg_confirmation))
gg_confirmation = np.asarray(gg_confirmation)
sg_confirmation = np.asarray(sg_confirmation)
gg_deviation = np.asarray(gg_deviation)
sg_deviation = np.asarray(sg_deviation)

ss_confirmation = np.asarray(ss_confirmation)
gs_confirmation = np.asarray(gs_confirmation)
ss_deviation = np.asarray(ss_deviation)
gs_deviation = np.asarray(gs_deviation)
```

```python
# G
x = range(1, 9)
fig, ax = plt.subplots()
ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)
plt.plot(x, gg_confirmation[0], "-", label="G8", color="k")
plt.plot(x, gg_confirmation[1], "--", label="G12", color="k")
plt.plot(x, gg_confirmation[2], ":", label="G16", color="k")
# plt.plot(x, gg_satisfaction_across_knowledge[3][1:], "-", label="G16", color=morandi_red)
plt.xlabel('K', fontweight='bold', fontsize=12)
plt.ylabel('Joint Confirmation', fontweight='bold', fontsize=12)
# plt.title("GG")
plt.xticks(x)
plt.legend(frameon=False, ncol=1, fontsize=12)
plt.savefig(data_folder + r"\Joint_Confirmation_across_Knowledge_GG.png", transparent=True, dpi=300)
plt.show()
```

```python
# Average it across K
gg_confirmation_avg = gg_confirmation.mean(axis=1)
sg_confirmation_avg = sg_confirmation.mean(axis=1)
joint_confirmation_avg_G = (gg_confirmation_avg + sg_confirmation_avg) / 2

gg_deviation_avg = gg_deviation.mean(axis=1)
sg_deviation_avg = sg_deviation.mean(axis=1)
mutual_deviation_avg_G = (gg_deviation_avg + sg_deviation_avg) / 2

x = [8, 12, 16]
fig, ax = plt.subplots()
ax.spines["left"].set_linewidth(1.5)
ax.spines["right"].set_linewidth(1.5)
ax.spines["top"].set_linewidth(1.5)
ax.spines["bottom"].set_linewidth(1.5)
plt.plot(x, joint_confirmation_avg_G, "-o", label="Joint Confirmation", color="k")
plt.plot(x, mutual_deviation_avg_G, "--v", label="Mutual Deviation", color="k")
plt.xlabel('Knowledge', fontweight='bold', fontsize=12)
plt.ylabel('Collective Interpretations', fontweight='bold', fontsize=12)
plt.title("Generalist")
plt.xticks(x)
plt.legend(frameon=False, ncol=1, fontsize=12)
plt.savefig(data_folder + r"\Generalist_interpretation.png", transparent=True, dpi=300)
plt.show()
```
