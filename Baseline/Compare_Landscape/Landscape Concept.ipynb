```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the landscape function with a single global optimal peak
def landscape_function(x, y):
    return np.sin(np.sqrt(x**2 + y**2)) + 0.5 * np.sin(np.sqrt((x - 2)**2 + (y - 2)**2))

# Generate data for the landscape surface
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = landscape_function(X, Y)

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface using grey colormap
surf = ax.plot_surface(X, Y, Z, cmap='gray')

# Hide the axes
ax.set_axis_off()

# Show the plot
plt.savefig("GreyLandscape.png", transparent=True, dpi=300)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the broader and lower Gaussian-like function with one peak
def broad_low_gaussian_surface(x, y):
    sigma_x = 2.0
    sigma_y = 3.0
    return np.exp(-(x**2 / (2 * sigma_x**2) + y**2 / (2 * sigma_y**2)))

# Generate data for the broader and lower Gaussian surface
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = broad_low_gaussian_surface(X, Y)

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the broader and lower Gaussian surface with a colormap (you can use any other colormap)
surf = ax.plot_surface(X, Y, Z, cmap='viridis', linewidth=0, antialiased=True)

# Hide the axes
ax.set_axis_off()

# Show the plot
plt.savefig("SmoothLandscape.png", transparent=True, dpi=300)
plt.show()
```
